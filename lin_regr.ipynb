_Моделирование данных и применение линейной регрессии_ 

Сгенерировал простые данные (полином второго порядка). Применил метод линейной регрессии. Далее применил полиномилаьную регрессию.

Почему-то коэфициенты дают не те что надо


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
x=np.random.rand(100)
y=x*x+2*x+1+(np.random.rand(100)-0.5)*0.01
```

```python
x=x.reshape(100,1)
y=y.reshape(100,1)

x_r=np.linspace(0,1,100).reshape(100,1)
y_r=(x_r*x_r+2*x_r+1).reshape(100,1)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x,y)
plt.plot(x_r,y_r,c="r")
```

```python
lr=LinearRegression()
lr.fit(x,y)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x,y)
plt.plot(x_r,y_r,c="r")
plt.plot(x_r,lr.predict(x_r),c="g")
```

```python
poly=PolynomialFeatures(degree=2)
```

```python
x_poly=poly.fit_transform(x)
```

```python
lr2=LinearRegression()
lr2.fit(x_poly,y)
```

```python
a=lr2.coef_
```

```python
a[0][0], a[0][1], a[0][2]
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x,y)
#plt.plot(x_r,y_r,c="r")
plt.plot(x_r, a[0][2]*x_r*x_r+ a[0][1]*x_r+a[0][0],c="g")

plt.plot(x_r, lr2.predict(poly.fit_transform(x_r)),c="r")
```
